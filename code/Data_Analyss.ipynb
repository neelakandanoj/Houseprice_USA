```python
import numpy as np
import pandas as pd
```

```python
df_CS = pd.read_csv(r"E:\House LCC\Datasets\CSUSHPISA.csv")

#Changing dtype of date column
df_CS["DATE"] = pd.to_datetime(df_CS["DATE"])

#Selecting data post JUNE 2001
mask = df_CS["DATE"] >= "2001-07-01"
df_CS = df_CS[mask]

#Resetting Index
df_CS.reset_index(inplace = True)
df_CS.drop(columns = ["index"], inplace = True)

#Creating "Year" and "Month" columns
df_CS['Year']=pd.DatetimeIndex(df_CS['DATE']).year
df_CS['Month']=pd.DatetimeIndex(df_CS['DATE']).month
print(df_CS.shape)
df_CS.tail()
```

```python
#Reading Unemp data
df_unemp=pd.read_csv(r'E:\House LCC\Datasets\UNRATE.csv')
mask = df_unemp["DATE"] >= "2001-07-01"
df_unemp = df_unemp[mask]
df_unemp.reset_index(inplace = True)
df_unemp.drop(columns = ["index"], inplace = True)
df_unemp.drop([267],inplace=True)
print(df_unemp.shape)
df_unemp.tail()
```

```python
# Reading Per Capita GDP Data into a dataframe
df_pcgdp = pd.read_csv(r"E:\House LCC\Datasets\gdp_per_capita.csv", names = ["DATE", "Per_Capita_GDP"], skiprows = 1)
mask = df_pcgdp["DATE"] >= "2001-07-01"
df_pcgdp = df_pcgdp[mask]
df_pcgdp.reset_index(inplace = True)
df_pcgdp.drop(columns = ["index"], inplace = True)
print(df_pcgdp.shape)
df_pcgdp.tail()
```

The data is quarterly. We will impute for other months using linear interpolation after we create the final dataframe combining all the data.

```python
# Interest Rate Data
df_Fed_rate = pd.read_csv(r"E:\House LCC\Datasets\FEDFUNDS.csv")
mask = df_Fed_rate["DATE"] >= "2001-07-01"
df_Fed_rate = df_Fed_rate[mask]
df_Fed_rate.reset_index(inplace = True)
df_Fed_rate.drop([267],inplace=True)
df_Fed_rate.drop(columns = ["index"], inplace = True)
print(df_Fed_rate.shape)
df_Fed_rate.tail() 
```

```python
df_cons_price_index = pd.read_csv(r"E:\House LCC\Datasets\Construction_Price.csv", names = ["DATE", "Cons_Materials"], skiprows = 1)
# df_cons_price_index.drop([252], inplace = True)
mask = df_cons_price_index["DATE"] >= "2001-07-01"
df_cons_price_index = df_cons_price_index[mask]
df_cons_price_index.reset_index(inplace = True)
df_cons_price_index.drop(columns = ["index"], inplace = True)
print(df_cons_price_index.shape)
df_cons_price_index.tail()
```

```python
df_CPI = pd.read_csv(r"E:\House LCC\Datasets\CPIAUCSL.csv", names = ["DATE", "CPI"], skiprows = 1).drop([252])
mask = df_CPI["DATE"] >= "2001-07-01"
df_CPI = df_CPI[mask]
df_CPI.reset_index(inplace = True)
df_CPI.drop(columns = ["index"], inplace = True)
print(df_CPI.shape)
df_CPI.tail()
```

```python
# Housing Subsidies

df_subsidy = pd.read_csv(r"E:\House LCC\Datasets\Housing_Subsidies.csv", names = ["DATE", "Subsidy"], skiprows = 1)
print(df_subsidy.shape)
df_subsidy.tail()
```

```python
# Number of households

df_households = pd.read_csv(r"E:\House LCC\Datasets\Households.csv", names = ["DATE", "Num_Households"], skiprows = 1)
print(df_households.shape)
df_households.tail()
```

```python
df_income = pd.read_csv(r"E:\House LCC\MedianIncome.csv", names = ["DATE", "Income"], skiprows = 1)
mask = df_income["DATE"] >= "2001-07-01"
df_income = df_income[mask]
df_income.reset_index(inplace = True)
df_income.drop(columns = ["index"], inplace = True)
print(df_income.shape)
df_income.tail()
```

```python
# Merging Per Capita GDP (Quarterly data)
df_pcgdp["DATE"] = pd.to_datetime(df_pcgdp["DATE"])
df_CS = pd.merge(df_CS,df_pcgdp, how = "left")
df_CS.head()
```

```python
# Concating dataframes having monthly data to create one dataframe
df = pd.DataFrame()
df_bymonth = [df_CS,df_CPI, df_unemp, df_cons_price_index, df_Fed_rate,df_income]
for df1 in df_bymonth:
    df1["DATE"] = pd.to_datetime(df1["DATE"])
    df1 = df1.set_index("DATE")
    df = pd.concat([df,df1], axis = 1)
print(df.shape)
df.head()
```

The "Per_Capita_GDP" column has missing values because the data was quarterly. The missing values in the other columns is due to unavailability of fresh data. We will first fill the missing values in the "Per_Capita_GDP" column using linear interpolation. We will drop the rows having missing values in the other columns. This means that we will use data from 2001 to 2020

```python
# Filling missing values in the Per_Capita_GDP column using linear interpolation
df["Per_Capita_GDP"] = df["Per_Capita_GDP"].interpolate()
```
